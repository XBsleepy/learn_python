```python
import re
# 处理多个分隔符号
line='asdf fjdk; afed,fjek,asdf, foo'
# 用分隔符可以是逗号，分号，或者空格，并且后面紧跟着任意个的空格
print(re.split(r'[;,\s]\s*',line))
# 用括号捕获分组,捕获的分隔符也会包含在结果中
fields=re.split(r'(;|,|\s)\s*',line)
print(fields)
# 非捕获分组,用(?:...)
fields=re.split(r'(?:;|,|\s)\s*',line)
print(fields)
# startswith()和endswith()方法，可以用元组来指定多个选项
filename='spam.txt'
print(filename.endswith('.txt'))
print(filename.startswith('file:'))
print(filename.startswith(('file:','spam:')))

# re的模式匹配，以日期为例子
# 复杂情况下，可以先将模式字符串编译成一个模式对象：
# 加()表示一个捕获组，可以用group()方法提取匹配的部分
datepat=re.compile(r'(\d+)/(\d+)/(\d+)')# d+代表匹配一个或多个数字
m=datepat.match('11/27/2012')
print(m.group(0))#group(0)是整个匹配，group(1)是第一个括号匹配的部分，group(2)是第二个括号匹配的部分...
print(m.groups())

# re.sub()函数可以用来替换字符串中的匹配项
text='Today is 11/27/2012. PyCon starts 3/13/2013.'
print(datepat.sub(r'\3-\1-\2',text))# \3表示第三个括号匹配的部分，\1表示第一个括号匹配的部分，\2表示第二个括号匹配的部分

# re.IGNORECASE标记，忽略大小写
text='UPPER PYTHON, lower python, Mixed Python'
print(re.findall('python',text,flags=re.IGNORECASE))
print(re.sub('python','snake',text,flags=re.IGNORECASE))

# str.translate()方法可以同时删除和替换字符
# 创建一个转换表格，然后用translate()方法
# str.maketrans()方法可以创建一个转换表格
# 用None表示删除字符
remap={ord('a'):'A',ord('b'):'B',ord('c'):'C'}
print(remap)
print('abc'.translate(remap))
print("".join(['a','b','c']))
```
